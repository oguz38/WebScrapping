```python
import yfinance as yf

tesla = yf.Ticker("TSLA")
tesla_stock_data = tesla.history(period="max")

print(tesla_stock_data.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find("table")
tesla_revenue = pd.read_html(str(table))[0]

print(tesla_revenue.head())
```

```python
gamestop = yf.Ticker("GME")
gamestop_stock_data = gamestop.history(period="max")

print(gamestop_stock_data.head())
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')


table = soup.find("table")
gamestop_revenue = pd.read_html(str(table))[0]

print(gamestop_revenue.head())
```

```python
gamestop_revenue.columns.tolist()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
plt.plot(tesla_stock_data.index, tesla_stock_data['Close'], label='Tesla Stock Price')
plt.title('Tesla Stock Price')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()


```

```python
plt.figure(figsize=(14, 7))
plt.plot(pd.to_datetime(tesla_revenue['Tesla Annual Revenue (Millions of US $)']), tesla_revenue['Tesla Annual Revenue (Millions of US $).1'], label='Tesla Revenue')
plt.title('Tesla Revenue')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

```python
# Plotting GameStop stock price
plt.figure(figsize=(14, 7))
plt.plot(gamestop_stock_data.index, gamestop_stock_data['Close'], label='GameStop Stock Price')
plt.title('GameStop Stock Price')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()

# Plotting GameStop revenue
plt.figure(figsize=(14, 7))
plt.plot(pd.to_datetime(gamestop_revenue['GameStop Annual Revenue (Millions of US $)']), gamestop_revenue['GameStop Annual Revenue (Millions of US $).1'], label='GameStop Revenue')
plt.title('GameStop Revenue')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```
